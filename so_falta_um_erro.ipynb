```python
from sklearn.cluster import KMeans
import numpy as np

#importando base de dados
df = pd.read_csv("MLiHops.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler

array1 = []

dicRecomendacao = {

  "Harmonizacao": {
  40 : "AMERICAN PALE ALE - Carnes grelhadas e saladas cítricas",
  2: "PALE LAGER - Hamburger, carnes vermelha e petiscos",
  43: "BROWN ALE - Carnes e petiscos",
  41: "WITBIER - Saladas, frutas do mar e comidas leves",
  44: "TRIPEL - Massas e frutos do mar",
  42: "BLONDE - Petiscos e frutos do mar",
  46: "STRONG - Frutos do mar e massas",
  45: "INDIAN PALE ALE - Queijo, carne vermelha, hamburger e petiscos",
  20: "FRUIT BEER - Sobremesa",
  },
  "Temperatura": {
  40: [8, 12],
  2: [0, 4],
  43: [8, 12],
  41: [4, 7],
  44: [4, 7],
  42: [2, 5],
  46: [10, 12],
  45: [8, 12],
  20: [7, 10],
  }
}

for i in range(25):  #itera sobre os valores e adiciona na lista array1

    temp = []
    temp.clear()

    p1 = df.iloc[i]['ibu']
    temp.append(p1)
    p2 = df.iloc[i]['abv (%)']
    temp.append(p2)
    p3 = df.iloc[i]['corpo']
    temp.append(p3)
    p4 = df.iloc[i]['aroma']
    temp.append(p4)
    p5 = df.iloc[i]['tipo']
    temp.append(p5)
    #p6 = dp.iloc[]["preco"]
    #temp.append(p6)

    array1.append(temp)

mms = MinMaxScaler()
X = np.array(array1) #transforma a lista array1 em numpy
mms.fit(X)
data_transformed = mms.transform(X)

print(array1)
```

```python
from sklearn.metrics import silhouette_score

num = 0
num1 = 0
for i in range(2, 15):
  km = KMeans(n_clusters=i, random_state=0)
  km.fit_predict(data_transformed)
  score = silhouette_score(data_transformed, km.labels_, metric='euclidean')
  if score > num:
    num = score
    num1 = i
  else:
    pass
kmeans = KMeans(n_clusters=num1, random_state=0).fit(X)
#pega todas a variaveis e as coloca em 10 grupos (mais parecidas)
tips = kmeans.labels_
print(tips)
```

```python
import decimal 

def busca(array,beer):
    for i,el in enumerate(array):
        if el==beer:
            return i
    return None

def arrendondar (x):
    return int(decimal.Decimal(x).quantize(0, decimal.ROUND_HALF_UP))

def sum_array(array):
  if len(array)==0:
    return None
  soma = 0
  for i in array:
    soma += i
  return(soma/len(array))

def ultimaOcorrencia(df, cerveja):
  array = df["Subrand"].to_numpy()
  #print(array)
  #print(df)
  cont = 0
  while cont < len(array):
    if array[cont] == cerveja:
      return cont

    cont +=1
  return None

################################################################################
def DataPreco(data, cerveja,conteiner,empdata,contador,quantidade_recomendada):
  mes = data[3:5]
  ano = data[6:]
  dia = data[:2]
  mes_atual = mes
  ano_atual = str(int(ano) - 1)
  dia_atual = dia
  dia30 = [4, 6, 9, 11]
  quantidade_container = 0
  empdata.sort_values(by=['Doc. Date'], inplace = True, ascending = False)
  #df = empdata.loc[(empdata['Doc. Date'] >= f'{dia_atual}/{mes_atual}/{ano_atual}') & (empdata['Doc. Date'] < f'{dia_atual}/{mes_atual}/{ano}')]
  df = empdata.loc[empdata['Unidade Container']==conteiner]
  litros = df.iloc[0]['Container Size 2']
  ult = ultimaOcorrencia(df, cerveja)
  ult_data = df.iloc[ult]["Doc. Date"]
  dianovo = ult_data.day
  mesnovo = ult_data.month
  anonovo = ult_data.year
  primeirodia = dianovo
  primeiromes = mesnovo - 1
  primeiroano = anonovo
  if primeiromes == 0:
    primeiromes = 12
    primeiroano = anonovo - 1
  #print(ult_data)
  #print(filtered_df.head())
  #print(f'{primeirodia}/{primeiromes}/{primeiroano}')
  #print(f'{dianovo}/{mesnovo}/{anonovo}')
  df.loc[(df['Doc. Date'] >= f'{primeirodia}/{primeiromes}/{primeiroano}') & (df['Doc. Date'] <= f'{dianovo}/{mesnovo}/{anonovo}')]
  df = df.loc[df["Subrand"] == cerveja]
  data_array = df['Preço/L'].to_numpy()
  media = sum_array(data_array)
  if media != None:
      #print("{1 CAIXA COM %d UNID E %.2f LITROS\npreco medio do L : %.2f"%(conteiner,litros,media))
      #print("1 CAIXA saira por %.2f $ (media * litragem da caixa)\
      #\ncolocar na parte de baixo de uma maneira mais organizada}"%(litros * media))
      if contador == 0:
          print("Recomendamos que voce compre %.2f litros dessa cerveja:"%(quantidade_recomendada))
      
      quantidade_container = arrendondar(quantidade_recomendada/litros)
      print("%d) %d container de %.2f L com %d unidades, com preco media do Litro de %.2f\nDando um total de %.2f $"
            %((contador+1),quantidade_container,litros,conteiner, media, quantidade_container * litros * media))
  return

def funcao(data, cerveja,quantidade_recomendada, empdata):
    #fazer lista com os tipos de conteiner da cerveja 
    filtered_df = empdata.loc[empdata["Subrand"] == cerveja].copy()
    #print(filtered_df)
    array_available_conteiners = filtered_df['Unidade Container'].unique()
    
    #print(array_available_conteiners[0])
    num_conteiner = filtered_df['Unidade Container'].nunique()
    c = 0
    conteiner = 0
    while c<num_conteiner:
        DataPreco(data, cerveja,array_available_conteiners[c], empdata,c,quantidade_recomendada)
        c+=1
    
    #print("\nRecomendamos que voce compre T litros dessa cervaja:\n1) Y container de X\nDando um total de W $")
    #if True:
     #   print("2) YY container de XX\nDando um total de WW $")
    # preco_medio = 0
    return
```
