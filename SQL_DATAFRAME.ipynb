```python
#
#   ESSE PROGRAMA PEGA A TABELA DO MYSQL E CONVERTE EM UM DATAFRAME 
#

import mysql.connector as connection
import pandas as pd

try:
    # PRECISA SUBSTITUIR LOCALHOST PELO SERVER ADDRESS
    mydb = connection.connect(host='127.0.0.1', database = 'db_usuario',
                              user='root', passwd='joaoeduardomiguel',use_pure=True)
                                                      
    #parametros
    table_name = 'user_table'
    identifier = 'Postal_Code'
    user_wanted = 2500
    
    #command
    query = f'''Select * from {table_name} 
    WHERE {identifier}={user_wanted}
    ORDER BY {identifier} DESC;'''
    result_dataFrame = pd.read_sql(query,mydb)
    
    
    mydb.close() #close the connection
    
except Exception as e:#FAIL HANDLER
    mydb.close()
    print("ERROR - NAO FOI POSSIVEL LOCALIZAR O SERVIDOR")


#output    
display(result_dataFrame)
```
