```python
from sklearn.cluster import KMeans
import numpy as np

#importando base de dados
df = pd.read_csv("MLiHops.csv")
df
```

```python
array1 = []

dicRecomendacao = {

  "Harmonizacao": {
  2 : "AMERICAN PALE ALE - Carnes grelhadas e saladas cítricas",
  40: "PALE LAGER - Hamburger, carnes vermelha e petiscos",
  43: "BROWN ALE - Carnes e petiscos",
  41: "WITBIER - Saladas, frutas do mar e comidas leves",
  44: "TRIPEL - Massas e frutos do mar",
  42: "BLONDE - Petiscos e frutos do mar",
  45: "STRONG - Frutos do mar e massas",
  46: "INDIAN PALE ALE - Queijo, carne vermelha, hamburger e petiscos",
  20: "FRUIT BEER - Sobremesa",
  },
  "Temperatura": {
  2: [8, 12],
  40: [0, 4],
  43: [8, 12],
  41: [4, 7],
  44: [4, 7],
  42: [2, 5],
  45: [10, 12],
  46: [8, 12],
  20: [7, 10],
  }
}

for i in range(25):  #itera sobre os valores e adiciona na lista array1

    temp = []
    temp.clear()

    p1 = df.iloc[i]['ibu']
    temp.append(p1)
    p2 = df.iloc[i]['abv (%)']
    temp.append(p2)
    p3 = df.iloc[i]['corpo']
    temp.append(p3)
    p4 = df.iloc[i]['aroma']
    temp.append(p4)
    p5 = df.iloc[i]['tipo']
    temp.append(p5)
    #p6 = dp.iloc[]["preco"]
    #temp.append(p6)

    array1.append(temp)

X = np.array(array1) #transforma a lista array1 em numpy

print(array1)
```

```python
kmeans = KMeans(n_clusters=9, random_state=0).fit(X)
#pega todas a variaveis e as coloca em 10 grupos (mais parecidas)
tips = kmeans.labels_
print(tips)
```
